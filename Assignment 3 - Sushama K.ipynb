# 3. Who scores the most on average for math, reading and writing based on
 Gender
 Test preparation course

```python
mean=df.groupby(['gender', 'test preparation course'])[['math score', 'reading score', 'writing score']].mean()
print(mean)
```

```python
import seaborn as sns
plt.figure(figsize=(15,5))
plt.title('Math Scores')
ax= sns.barplot(x='test preparation course',y='math score',hue= 'gender',data=df,errwidth=0)  
plt.xlabel('test preparation course')
plt.ylabel('math score')
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Reading Scores')
ay= sns.barplot(x='test preparation course',y='reading score',hue= 'gender',data=df,errwidth=0)  
plt.xlabel('test preparation course')
plt.ylabel('reading score')
for i in ay.containers:
    ay.bar_label(i,)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Writing Scores')
az= sns.barplot(x='test preparation course',y='writing score',hue= 'gender',data=df,errwidth=0)  
plt.xlabel('test preparation course')
plt.ylabel('Writing score')
for i in az.containers:
    az.bar_label(i,)
plt.show()
```

# From the bar plot it is evident that the person who completed the test preparation course have scored more than those who didnot do the test preparation course. And among female and male, female scored more than male in writing and reading ,but in the case of math score male scored more than female

# 4. What do you think about the scoring variation for math, reading and writing based on
 Gender  Test preparation course


```python
df.groupby(['gender', 'test preparation course'])[['math score', 'reading score', 'writing score']].std()
```

```python
df4 = pd.DataFrame({
    'gender': ['female', 'female', 'male', 'male'],
    'test preparation course': ['completed', 'none', 'completed', 'none'],
    'math score': [14.428233, 16.510686, 14.039164, 14.335746],
    'reading score': [12.695493, 14.636981, 13.612420,13.595718],'writing score': [12.240788, 15.016497, 12.887910, 13.468284]})
 
print(df4)
```

```python
plt.figure(figsize=(15,5))
plt.title('Math Scores')
uv= sns.barplot(x='test preparation course',y='math score',hue= 'gender',data=df4,errwidth=0)  
plt.xlabel('test preparation course')
plt.ylabel('math score')
for i in uv.containers:
    uv.bar_label(i,)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Writing Scores')
uw= sns.barplot(x='test preparation course',y='writing score',hue= 'gender',data=df4,errwidth=0)  
plt.xlabel('test preparation course')
plt.ylabel('Writing score')
for i in uw.containers:
    uw.bar_label(i,)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Reading Scores')
ux= sns.barplot(x='test preparation course',y='reading score',hue= 'gender',data=df4,errwidth=0)  
plt.xlabel('test preparation course')
plt.ylabel('Reading score')
for i in ux.containers:
    ux.bar_label(i,)
plt.show()
```

# From the bar plots, it is clear that, In case of math score female have more variation than males for both the groups who have or not have attended the test preparation course. In case Reading and Writng score, In the case of people who have completed course, males have more variation than females and in the case people who didnt attend the course, females show more variation than males.
